```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('match_result_data.csv')
df.head()
```

```python
df['SEASON'].unique()
```

```python
years = [2017, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2018,2019]
years.sort()      
```

```python
if 'TOSS_WINNER_MODEL' in df.columns:
    df.drop(['TOSS_DECISION_MODEL'], axis=1, inplace = True)
```

```python
if 'WINNER_MODEL' in df.columns:
    df.drop(['WINNER_MODEL'], axis=1, inplace = True)
```

```python
if 'TOSS_DECISION_MODEL' in df.columns:
    df.drop(['TOSS_DECISION_MODEL'], axis=1, inplace = True)
```

```python
Toss_result = pd.Series([]) 
Winner_result = pd.Series([])
Toss_deision_result = pd.Series([])
for(i,row) in df.iterrows():
    
    if row["TOSS_WINNER"] == row["TEAM2"]:
        toss_result = 0
    else:
        toss_result = 1
    Toss_result[i]= toss_result
    
    if row["WINNER"] == row["TEAM2"]:
        winner_result = 0
    else:
        winner_result = 1
        
    if row["TOSS_DECISION"] == 'field':
        toss_decision_result = 0
    else:
        toss_decision_result = 1
        
    Toss_result[i]= toss_result
    Winner_result[i]= winner_result
    Toss_deision_result[i]= toss_decision_result
    
df.insert(9, "TOSS_WINNER_MODEL", Toss_result) 
df.insert(13, "WINNER_MODEL", Winner_result) 
df.insert(11, "TOSS_DECISION_MODEL", Winner_result) 
df.head() 
```

# The Below Code is to make different csv files for each season

```python
for i,year in enumerate(years):
    print(i,year)
    new_df = df.loc[df['SEASON'] == year]
    new_df.reset_index(drop = True)
    filename = "IPL_" + str(year) + ".csv"
    new_df.to_csv(filename, index=False)
```
